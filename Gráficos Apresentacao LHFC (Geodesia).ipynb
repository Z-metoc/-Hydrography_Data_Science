```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Desktop/desvio_final")
df
```

```python
#desvio padrao e Quality maximo
```

```python
fig,ax=plt.subplots(figsize=(12,9))
ax.bar(df.index,df["Q"],color=(0.1, 0.1, 0.1,0.1),label="Quality",edgecolor="b")
ax.legend(loc="upper center")
ax.set_ylabel
ax2 = ax.twinx()

ax2.plot(df.index,df["SDHeig"],label="Desvio Padrão Horizontal",linestyle='dashed',color="y")
ax2.plot(df.index,df["SDHori"],label="Desvio Padrão Vertical",linestyle='dashed', color="r")
ax2.plot(30)

ax.set_ylabel("QUALIDADE")
ax2.set_ylabel("DESVIO PADRÃO")
ax.set_xlabel("TEMPO(Segundos)")

plt.show()
```
