```python

import numpy as np
import matplotlib.pyplot as plt  # importando libs
```

```python
     # DESENVOLVIMENTO GRAFICO - ATIVIDADE MARE ESTATICA#

t=np.linspace(0,4.4*np.pi,80) # em radiano , espaçado de 10 em 10 , 800 graus / 10 = 80 de espaçamento
mare=0.44*np.cos(2*(t)) - 0.351 # resultado  lat= 0/dec= 25N
mare_desi = 0.038 + 0.215*np.cos(2*(t)) + 0.070*np.cos(t) # resultado  lat=10S/dec= 25N
mare_diurna = 0.000546 + 0.1489*np.cos(2*(t)) + 0.1945*np.cos(t) # resultado  lat=35S/dec=25N
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(t,mare,linestyle='dashed',label=" lat=0/dec=25N ")
ax.plot(t,mare_desi,linestyle="dashed",color="r",label="lat=10S/dec=25N")
ax.plot(t,mare_diurna,linestyle='dashed',color="k",label="lat=35S/dec=25N")
ax.set_ylabel("AMPLITUDE")
ax.set_xlabel("VARIACAO ANGULO HORARIO (RADIANOS)")
ax.set_title("GRAFICO MARE ESTATICA")
ax.legend()
plt.show()
```
