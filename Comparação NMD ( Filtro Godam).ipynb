
    # Comparação NMD de Estações Maregaraficas impletado através do Filtro GODAM #

```python
# importando bibliotecas 
import numpy as np 
import matplotlib.pyplot as plt
from pandas import read_csv
from matplotlib import pyplot
```

```python
# importando dados de maré utilizados e criando dataframes
cadim= "/Users/thiagozamith/Desktop/cadim"
cadim=read_csv(cadim,parse_dates=False)

ilha_bernardo = "/Users/thiagozamith/Desktop/Ilha_Bernardo"
ilha_bernardo=read_csv(ilha_bernardo,parse_dates=False)

cadim
```

```python
#CADIM
# Implementando Filtro Godam para comparação do Nivel Medio Diario (NMD) - Eliminando M2
media_movel01=cadim.rolling(window=25)
media_movel25h=media_movel01.mean()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(cadim.index,cadim["Altura(cm)"])
ax.plot(media_movel25h.index,media_movel25h["Altura(cm)"],color="red")
ax.set_title("Comparação Fitro 25h-M2 x Ilha do Bernardo")
ax.set_xlabel("Horas")
ax.set_ylabel("Altura(cm)")
plt.show()
```

```python
# Implementando Filtro Godam para comparação do Nivel Medio Diario (NMD) - Eliminando S2
media_movel02=media_movel25h.rolling(window=49)
media_movel24h=media_movel02.mean()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(cadim.index,cadim["Altura(cm)"])
ax.plot(media_movel24h.index,media_movel24h["Altura(cm)"],color="red")
ax.set_title("Comparação Fitro 24h-S2 x Ilha do Bernardo")
ax.set_xlabel("Horas")
ax.set_ylabel("Altura(cm)")

plt.show()
```

```python
# Implementando Filtro Godam para comparação do Nivel Medio Diario (NMD) - Eliminando Residuos M2/S2
media_movel03=media_movel24h.rolling(window=74)
media_movel25h_res=media_movel03.mean()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(cadim.index,cadim["Altura(cm)"])
ax.plot(media_movel25h_res.index,media_movel25h_res["Altura(cm)"],color="red")
ax.set_title("Comparação Fitro 25h-Residuos x CADIM")
ax.set_xlabel("Horas")
ax.set_ylabel("Altura(cm)")


plt.show()
```
