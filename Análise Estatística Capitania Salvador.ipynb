```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#import plotly.express as px
%matplotlib inline
plt.rcParams['figure.figsize'] = (11,7)
```

```python
df=pd.read_csv("/Users/thiagozamith/Desktop/teste2") # importando o arquivo
print(df) # amostrando dados organizados
x2=df[["OBSERVADA"]] 
y2=df[["PREVISTA"]]
x2_=np.array(x2)
y2_=np.array(y2)
```

```python
plt.scatter(y2,x2)
```

```python
from sklearn.linear_model import LinearRegression
# criar modelo linear e otimizar
lm_model = LinearRegression()
lm_model.fit(x2_,y2_)
# extrair coeficientes
alfa = lm_model.coef_         # Caso ajustassemos para Yˆ=βˆ + β1 * x # uma reta #
beta = lm_model.intercept_

```

```python
plt.scatter(x2_, y2_,s=20)
plt.plot(x2_, (x2_ * alfa + beta), color='r')
plt.title('Regressao  Linear') 
plt.xlabel('OBSERVADA MORROSP') 
plt.ylabel('PREVISTA MORROSP') 
plt.show()


```

```python
import statsmodels.api as sm # lib estatistica

# é necessário adicionar uma constante a matriz X
x_sm = sm.add_constant(x2)
# OLS vem de Ordinary Least Squares e o método fit irá treinar o modelo
results = sm.OLS(y2, x_sm).fit()
# mostrando as estatísticas do modelo
results.summary()
```
